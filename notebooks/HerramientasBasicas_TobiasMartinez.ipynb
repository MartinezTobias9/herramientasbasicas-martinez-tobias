## 6. Limpieza y estructuración

### Duplicados

```python
df.drop_duplicates(inplace=True) #Eliminamos los registros duplicados. Con el parámetro inplace lo hacemos de manera permanente.
```

### Estandarización de fechas, categorias y unidades.

```python
df['Price'] = df['Price'].astype(str).str.replace(',', '') #Quitamos las comas
df['Price'] = pd.to_numeric(df['Price'], errors='coerce') #Aplicamos formato númerico
df.tail(8) #Vista rápida para ver como va quedando
```

A su vez, solo utilizaremos el año de la columna *Date*. Para ello, primero debemos darle formato de fecha y luego crear una columna llamada *Year*, donde se albergará unicamente el año de la misión.

```python
# Convertimos el texto a formato fecha
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Creamos la columna año
df['Year'] = df['Date'].dt.year
df.tail(8)
```

### Registros nulos

```python
#Eliminamos los registros vacios de la columna Price. Con el parámetro inplace lo hacemos de manera permanente.
df.dropna(subset=['Price'], inplace=True)
print(df.isnull().sum())
```

### Eliminación de columnas

```python
# Eliminamos la columnas que no son necesarias.
df.drop(columns=['Time', 'Rocket', 'Mission', 'RocketStatus'], inplace=True)# Con el parámetro inplace lo hacemos de manera permanente.
```

###1) ¿Se ha vuelto el acceso al espacio más económico con el desarrollo de nuevas tecnologías?

```python
# 1. Preparar datos con MEDIANA en lugar de PROMEDIO
missions_trend = df.groupby('Year').size().reset_index(name='Misiones')
# Utilizamos la función .median() para capturar la mediana
median_price_year = df.groupby('Year')['Price'].median().reset_index(name='Precio_Mediano')
combined_df = pd.merge(missions_trend, median_price_year, on='Year', how='left')
# 2. Variable numérica para la tendencia
x = np.arange(len(combined_df))
# Tendencia de las misiones
coef_misiones = np.polyfit(x, combined_df['Misiones'], 1)
trend_misiones = np.poly1d(coef_misiones)(x)
# Tendencia del precio
coef_precio = np.polyfit(x, combined_df['Precio_Mediano'], 1)
trend_precio = np.poly1d(coef_precio)(x)
# 3. Gráfico
fig = make_subplots(specs=[[{"secondary_y": True}]])
# Precio
fig.add_trace(
    go.Scatter(
        x=combined_df['Year'],
        y=combined_df['Precio_Mediano'],
        name="Mediana de Precio (MDD)",
        line=dict(color='rgba(100, 149, 237, 1)')
    ),
    secondary_y=True,
)
# Tendencia precio
fig.add_trace(
    go.Scatter(
        x=combined_df['Year'],
        y=trend_precio,
        name="Tendencia Precio (Mediana)",
        line=dict(color='rgba(100, 149, 237, 1)', dash='dash')
    ),
    secondary_y=True,
)
# Misiones reales
fig.add_trace(
    go.Scatter(
        x=combined_df['Year'],
        y=combined_df['Misiones'],
        name="Cantidad de Misiones",
        mode='lines+markers',
        line=dict(color='white')
    ),
    secondary_y=False,
)
# Tendencia misiones
fig.add_trace(
    go.Scatter(
        x=combined_df['Year'],
        y=trend_misiones,
        name="Tendencia Misiones",
        line=dict(color='white', dash='dash')
    ),
    secondary_y=False,
)
# Layout ajustado
fig.update_layout(
    title_text="Evolución y Tendencias: Volumen vs. Mediana de Costos",
    template='plotly_dark',
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)
fig.update_xaxes(title_text="Año")
fig.update_yaxes(title_text="Cantidad de Misiones", secondary_y=False)
fig.update_yaxes(title_text="Mediana de Precio (Millones USD)", secondary_y=True)
fig.show()
```

####Interpretación breve de Evolución de Misiones y su Costo Promedio (MDD)
Basado en el análisis, podemos distinguir una tendencia alcista en el volumen de misiones y una marcada reducción del costo en los últimos años, lo que refleja no solo un abaratamiento real gracias a la reutilización de cohetes, otras tecnologías y el ingreso del sector privado.

###2) ¿Existe una relacion entre el presupuesto invertido y el resultado de la misión?

```python
# Creamos el gráfico de dispersión de los costos por resultado de misiones.
fig = px.scatter(
    df,
    x='MissionStatus',
    y='Price',
    color='MissionStatus',
    title='Relación entre Presupuesto (MDD) y Resultado de la Misión',
    labels={'MissionStatus': 'Estado de la misión', 'Price': 'Precio (Millones USD)'},
    hover_data=['Company', 'Year'],
    template='plotly_dark'
)

# Ajustamos la opacidad y el tamaño
fig.update_traces(marker=dict(size=20, opacity=0.6))

fig.show()
```

####Interpretación breve de la relación entre presupuestos y resultados
A partir de la mediana podemos entender que las misiones exitosas tienden a tener un mayor presupuesto que las fallidas.

* Exitosas: 64.68 MDD
* Fallas parciales: 47 MDD
* Fallidas: 37 MDD
* Fallidas antes del despegue: 62 MDD (Solo 1 caso, baja representatividad)

A mayor inversión, mayor nivel de pruebas, validaciones y redundancias. Esto reduce el riesgo de fallo, sin embargo, **no lo garantiza**.

###3) ¿Existe un cambio en la estructura del sector espacial en términos de participación pública y privada?

```python
# Nueva columna llamada Decade agrupando los años por década
df['Decade'] = (df['Year'] // 10) * 10

# Agrupar por década y empresa, contando la cantidad de misiones
top_companies_by_decade = df.groupby(['Decade', 'Company']) \
                            .size() \
                            .reset_index(name='Misiones')
# Ordenar por década y cantidad de misiones (de mayor a menor)
# Luego tomar las 5 empresas con más misiones en cada década
top_5_by_decade = top_companies_by_decade \
                    .sort_values(['Decade', 'Misiones'], ascending=[True, False]) \
                    .groupby('Decade') \
                    .head(5)
# Gráfico de barras
fig = px.bar(
    top_5_by_decade,
    x='Decade',
    y='Misiones',
    color='Company',
    title='Top 5 Empresas con más Misiones por Década',
    barmode='group',  # Barras agrupadas por década
    template='plotly_dark'
)
fig.show()
```

####Interpretación breve de la reestructuracion del sector espacial
El gráfico muestra una tendencia al cambio progresivo en la composición del sector. En las primeras décadas dominaban actores gubernamentales como la NASA, RVSN USSR/ROSCOSMOS. A partir de 1990 aparecen mas competidores y desde los años 2000 crece con fuerza el sector privado, con empresas como SpaceX o Arianespace. Sin embargo, las agencias estatales no pierden peso, como lo evidencia el CASC, donde aumenta exponencialmente su participación.
